# Plot the famous Activation functions

```python
import matplotlib.pyplot as plt
import numpy

x = numpy.linspace(-10, 10, 1000)
```

```python
def sigmoid(x):
    return 1 / (1 + numpy.exp(-x))
def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, sigmoid_derivative(x), label='Sigmoid Derivative')
plt.legend()
plt.title('Sigmoid Activation Function and its Derivative')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```

```python
def tanh(x):
    return numpy.tanh(x)
def tanh_derivative(x):
    return 1 - numpy.tanh(x)**2

plt.plot(x, tanh(x), label='Tanh')
plt.plot(x, tanh_derivative(x), label='Tanh Derivative')
plt.legend()
plt.title('Tanh Activation Function and its Derivative')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```

```python
def relu(x):
    return numpy.maximum(0, x)
def relu_derivative(x):
    return numpy.where(x > 0, 1, 0)

plt.plot(x, relu(x), label='ReLU')
plt.plot(x, relu_derivative(x), label='ReLU Derivative')
plt.legend()
plt.title('ReLU Activation Function and its Derivative')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```

```python
def elu(x, alpha=1.0):
    return numpy.where(x >= 0, x, alpha * (numpy.exp(x) - 1))
def elu_derivative(x, alpha=1.0):
    return numpy.where(x >= 0, 1, elu(x, alpha) + alpha)

plt.plot(x, elu(x), label='ELU')
plt.plot(x, elu_derivative(x), label='ELU Derivative')
plt.legend()
plt.title('ELU Activation Function and its Derivative')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```

```python
def selu(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * numpy.where(x >= 0, x, alpha * (numpy.exp(x) - 1))
def selu_derivative(x, lambda_=1.0507, alpha=1.67326):
    return lambda_ * numpy.where(x >= 0, 1, alpha * numpy.exp(x))

plt.plot(x, selu(x), label='SELU')
plt.plot(x, selu_derivative(x), label='SELU Derivative')
plt.legend()
plt.title('SELU Activation Function and its Derivative')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid()
plt.show()
```
